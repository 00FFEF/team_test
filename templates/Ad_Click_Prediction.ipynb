```python
import pandas as pd
import sqlite3
```

```python
conn = sqlite3.connect("../db.sqlite3")
```

```python
df = pd.read_sql_query('select * from dbapp_admachine da', conn)
df = df.iloc[0:,1:]
df.info()
```

```python
conn.close()
```

```python
import matplotlib.pyplot as plt
```

```python
df_gen = df['Gender']
```

```python
male = []
female = []

for i in range(len(df_gen)):
  g = df_gen[i]
  if g == 'Male':
    male.append(g)
  else :
    female.append(g)

x = ['Male', 'Female']
y = [len(male), len(female)]

bar_colors = ['blue', 'red']
plt.bar(x, y, width = 0.3, color = bar_colors, edgecolor = 'black', linewidth = 1, align = 'center')

# 위에 도수를 나타내는 식
for i, v in enumerate(x):
    plt.text(v, y[i], y[i],
            fontsize = 15, color="magenta", horizontalalignment='center', verticalalignment='bottom')
plt.show()
```

# 추가할 것
* 막대 간격 좁히기
* 막대 그래프 숫자 표시 (o)

```python
df_age = df['Age']
```

```python
juv = []
tw = []
th = []
fo = []
fi = []
ot = []

for i in range(len(df_age)):
  n = df_age[i]
  if n < 20 :
    juv.append(n)
  elif n < 30 :
    tw.append(n)
  elif n < 40 :
    th.append(n)
  elif n < 50 :
    fo.append(n)
  elif n < 60:
    fi.append(n)
  else :
    ot.append(n)
```

```python
ratio = [len(juv), len(tw), len(th), len(fo), len(fi), len(ot)]
labels = ['Teenagers', 'Twenties', 'Thirties', 'Fouries', 'Fifties', 'Sixties']
wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 2}

title_font = {
    'fontsize': 20,
    'fontweight': 'bold'
}
plt.title('Generations', fontdict=title_font)
plt.legend(labels = labels, loc = 'upper right')
plt.pie(ratio, labels=labels, autopct='%d%%', shadow = True, wedgeprops=wedgeprops, startangle=90, textprops = {'fontsize':14})
plt.show
```
